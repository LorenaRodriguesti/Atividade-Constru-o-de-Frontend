```python
# Importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregando o dataset Iris direto da internet
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df = pd.read_csv(url)

# Exibir as 5 primeiras linhas
print("5 primeiras linhas do dataset:")
print(df.head())

# Gráfico 1: Comprimento x Largura das pétalas, colorido por espécie
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x="petal_length", y="petal_width", hue="species")
plt.title("Comprimento x Largura das Pétalas por Espécie")
plt.savefig("grafico_petalas.png")
plt.show()

# Gráfico 2: Boxplot do comprimento das sépalas
plt.figure(figsize=(10,6))
sns.boxplot(x="species", y="sepal_length", data=df)
plt.title("Distribuição do Comprimento das Sépalas por Espécie")
plt.savefig("boxplot_sepal_length.png")
plt.show()

# Gráfico 2 (continuação): Boxplot da largura das sépalas
plt.figure(figsize=(10,6))
sns.boxplot(x="species", y="sepal_width", data=df)
plt.title("Distribuição da Largura das Sépalas por Espécie")
plt.savefig("boxplot_sepal_width.png")
plt.show()

# Gráfico 3: Matriz de correlação
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Matriz de Correlação do Dataset Iris")
plt.savefig("matriz_correlacao.png")
plt.show()

# Respostas às perguntas
print("\nAnálise e respostas:")

print("a) Existe algum padrão claro que diferencia as espécies?")
print("- Sim, principalmente no gráfico de pétalas, as espécies Setosa, Versicolor e Virginica formam grupos distintos.")

print("\nb) Quais atributos parecem ser mais úteis para separar as classes?")
print("- Comprimento e largura das pétalas parecem ser os atributos mais relevantes para distinguir as espécies.")

print("\nc) Há sobreposição entre as espécies em algum gráfico?")
print("- Sim, nas sépalas, especialmente em largura, há uma certa sobreposição entre Versicolor e Virginica.")

print("\nd) Se você fosse treinar um modelo, quais colunas escolheria como mais relevantes? Por quê?")
print("- Eu escolheria as colunas 'petal_length' e 'petal_width' pois apresentam separação mais clara entre as espécies, facilitando a classificação do modelo.")
```
